# Working with Data in Python

## Importing Libraries

To work with data, we'll first need to import some libraries. Libraries are basically packages of code that someone has already written and made available for you to use.

A major library that we'll be using is called **pandas**. Pandas allows us to structure data in a tabular format (rows and columns) similar to Excel. Once we import it, we'll also give it an alias. The alias allows us to call pandas by a shorter name in our code.

```python
import pandas as pd
```

## Loading Data with Pandas

Now that we've imported the library, we have access to all the cool things pandas has to offer. The flagship data class that pandas offers is the **DataFrame**. The DataFrame is basically a table. It has rows, columns, and the columns have headers (column names). 

Let's import some data from NYC Open Data and start working with it. Today we'll be using data from **Citibike**, NYC's bike share program. Citibike collects data on every bike ride within the system and makes the data publicly available for people to use. 

We'll import the Citibike data from the month of September 2019 using pandas. We commonly use the variable name `df` for dataframes.

```python
url='https://raw.githubusercontent.com/nickxscott/INFO658/refs/heads/main/notebooks/data/citibike_201909.csv'

df = pd.read_csv(url)

#we use the .head() method on a dataframe to just view the first 5 rows of data.
df.head()
```

Let's see how many rows are in our dataframe using the `len()` function:

### Viewing Column Names

49,244 rows! That's a lot of data. Let's continue inspecting the dataset. We can see what all of the column names are using the `.columns` method. This returns all the column names in a list, which might be useful for some tasks we want to do later.

```python
df.columns
```

### Understanding Data Types

Let's see the different datatypes in each column using the `.dtypes` method. This will list each column name along with the datatype of the column. This will be helpful to know before we start doing any analysis.

```python
df.dtypes
```

## Accessing Individual Columns

We can view one column at a time by calling a column header in square brackets after our dataframe variable name. This will return a single column in an array (similar to a list). Let's look at the `bikeid` column:

### Shortcut for Accessing Columns

**Hint:** If your column name has no spaces in it, you can call it more easily by simply appending a period and the column name to the dataframe:

## Converting Data Types

We noticed that `bikeid` is being read as an integer (data type int64). Since this number is intended to be used as a unique identifier for a bike, and not an actual quantitative value that we can do math with, let's go ahead and change it to a string just to make sure we don't accidentally do any calculations on it. 

We'll cast the `bikeid` column as a string using the `.astype()` method. When we do this, we need to make sure we assign the new array of values to the original column in the dataframe.

```python
df['bikeid']=df.bikeid.astype(str)

#lets view our datatypes again to make sure we did it correctly
df.dtypes
```

## Sorting Data

Great! `bikeid` is now being read as a string (or "object" in pandas parlance).

Something else we might want to do with our dataframe is to sort it. We notice there's a column for trip duration. This is the length of the trip in seconds. Perhaps we want to sort our dataframe starting with the longest trip to the shortest. 

We can do this using the `.sort_values()` method and selecting a column. The default behavior of this method is to sort in ascending order (smallest to largest), so to do the opposite we need to add the argument `ascending=False`. Let's view the top 10 longest rides for this month:

```python
df.sort_values(by='tripduration', ascending=False).head(10)
```

## Data Cleaning: Removing Outliers

Wow! We have three rides that lasted over a million seconds. Looking at the `starttime` and `stoptime` columns, we see that these rides lasted for several weeks. 

We're pretty sure people don't ride citibikes for that long, so this was clearly a technical mistake. Perhaps the user didn't dock the bike correctly, or the bike was stolen and never docked. Regardless, we should remove these values from our dataset before moving forward, as we can't do much with them, and these big numbers will throw off any analysis that we try to do. 

Maybe it's a good idea to remove any ride longer than 2 hours. Let's see how many there are in our dataset. To find these, we'll first calculate the number of seconds in 2 hours, then find these values using the `.loc[]` method.

```python
#calculate how many seconds are in two hours
two_hours=2*60*60

df.loc[df.tripduration>=two_hours]
```

### Creating a Clean Dataset

251 trips lasted over 2 hours. Out of nearly 50,000 trips, that's not so bad. Let's create a new dataset that we'll call `df_clean` that has these rides removed. Again, we'll use the `.loc[]` method, but this time we'll just find the rows where the rides were under 2 hours:

```python
df_clean=df.loc[df.tripduration<=two_hours]
```

## Calculating Summary Statistics

Great! Now let's get some summary statistics from our data. Perhaps we want to know the average trip duration. We can find this using the `.mean()` method on the column:

```python
df_clean.tripduration.mean()
```

### Converting Seconds to Minutes

595 seconds. Let's see if we can convert that to minutes. We'll do that by dividing the seconds by 60 and using the `int()` function to strip off the decimal values. We'll then find the seconds using the modulo operator, which will give us the remainder of the division. We'll then use the `round()` function to round the decimal to the nearest second:

```python
minutes=int(df_clean.tripduration.mean()/60)

seconds=round(df_clean.tripduration.mean()%60)

#now let's print the result
print('The average trip was ', minutes, ':', seconds)
```

## Creating Reusable Functions

Great! Converting seconds to minutes sounds like something we might want to do again. Let's write a function to do this faster next time.

This will take seconds as an input and return a string in the form of `'minutes:seconds'`.

```python
def seconds_to_minutes(sec):
    minutes=int(sec/60)
    seconds=round(sec%60)
    #if seconds are under 10, we'll need to add a leading 0
    if seconds <10:
        seconds='0'+str(seconds)
    return str(minutes)+':'+str(seconds)

#Let's use our function to find the longest trip duration in the dataset and convert it to minutes! We'll use the .max() method to do this.
max_trip = df_clean.tripduration.max()

print('Max trip in seconds - ', max_trip)
print('Max trip in minutes - ', seconds_to_minutes(max_trip))
```

### Understanding Skewed Data

It seems like we might have skew to our data, even after cleaning up some of those long trips. This makes sense, as a ride can't last less than 0 seconds, but there's no max limit to how long they can last. That sounds like a classic set-up for right-skewed data. 

Let's find the median trip duration, as that might be a better measure of our typical trip duration than the mean. We'll use the `.median()` method:

```python
median_trip = df_clean.tripduration.median()

print('Median trip in seconds - ', median_trip)
print('Median trip in minutes - ', seconds_to_minutes(median_trip))
```

## Finding Unique Values

That sounds more like it. The median trip is 6 minutes and 12 seconds. 

I see there is a column that lists the station name where the ride starts and ends. Let's see how many stations there are. Of course, a station can be listed more than once in this dataset, as many rides could have started or ended at the same place. 

To see each station listed a single time, we'll use the `.unique()` method. This will return each unique value in an array:

```python
df_clean['start station name'].unique()
```

### Counting Unique Values

Now we have all of our stations in a list. Let's see how many there are:

```python
stations = df_clean['start station name'].unique()
print('There are a total of',len(stations),'stations in the dataset')
```

## Aggregating Data with GroupBy

Got it. Now let's see how many rides started at each station. To do this, we'll need to do some aggregating. We can aggregate in pandas using the `.groupby()` method, then specifying what we want to do with the aggregated values. 

In this case, we want to group by the start station name, then count each time that station name appears:

```python
df_clean.groupby(['start station name'])['start station name'].count()
```

### Creating a DataFrame from Grouped Data

Easy! Now we can see how many trips started at each station. If we want to see this as a dataframe that we can continue working with, we can apply the `.reset_index()` method and give our aggregated column a name. This will return a dataframe that we can save as a new variable:

```python
df_stationCount=df_clean.groupby(['start station name'])['start station name'].count().reset_index(name='count')

#now let's sort by the count column to see our top 5 most popular stations
df_stationCount.sort_values(by='count', ascending=False).head()
```

### Finding the Bottom 5

How about our bottom 5 stations? We can just remove the `ascending=False` argument:

```python
df_stationCount.sort_values(by='count').head()
```

### Average Trip Duration by Station

Let's try another aggregation method. Why don't we see the average trip duration based on where the trip started from:

```python
df_stationDuration=df_clean.groupby(['start station name'])['tripduration'].mean().reset_index(name='mean_trip_duration')
df_stationDuration
```

### Stations with Longest Average Trips

Now let's see the stations where the longest average trips originated from:

```python
df_stationDuration.sort_values(by='mean_trip_duration', ascending=False).head()
```

## Iterating Through DataFrames

So far so good. Now I want to add a column to this dataset. I really like being able to see the trip duration in minutes. It would be helpful if I had that in a column.

To do this, I need to convert every value in the `tripduration` column to minutes using my function `seconds_to_minutes`, store the output somewhere, then create a new column with all of the results. 

I can do this by iterating (or looping) through my dataframe using the method `.iterrows()`. This method returns the index and each row as an array separately. To demonstrate how this method works, I'll just loop through the first five rows of the dataset and print some values:

```python
for index, row in df_clean.head().iterrows():
    print(index, row['start station name'], row['tripduration'])
```

### Creating a New Column with a Loop

Now that we see how `iterrows()` works, let's create our `tripduration_minutes` column. First, we'll create a list variable to hold the values of our new column. Then we'll iterate through the dataset and run our `seconds_to_minutes` function on every `tripduration` value and store it in the list. Finally, we'll create a new column with the values in the list:

```python
minutes_column=[]

for index,row in df_clean.iterrows():
    sec=row.tripduration
    minutes_string=seconds_to_minutes(sec)
    #now append the new value to the list above
    minutes_column.append(minutes_string)
#create a new column and set the values as our list
df_clean['tripduration_minutes']=minutes_column

#now let's inspect our work!
df_clean
```

### Handling the Warning

Pandas gave us a warning! It's telling us that we're creating a new column based on a "slice" of our original dataframe. This is because `df_clean` is just a subset of our original dataframe. You can just ignore this warning, or silence it by including the following code when you import pandas:

```python
pd.options.mode.chained_assignment = None
```

## Mapping Values with Dictionaries

Let's look at the column called `gender`. It looks like it's just a number. After reviewing the metadata, I found out that the values mean the following:
- 0 = unknown
- 1 = male  
- 2 = female

Let's convert these gender codes to the actual names using a dictionary and the `.map()` method, and we'll create a new column to store it.

When we use the `.map()` method on the gender column and pass it our dictionary, it treats the column values as keys of the dictionary and "maps" them to the corresponding values. This is an extremely useful way to create a new column from a category code of this nature.

```python
d_gender={0:'unknown',
          1:'male',
          2:'female'}

df_clean.gender.map(d_gender)
```

### Creating the Gender Description Column

Now let's make a new column to show the name of the gender of the rider:

```python
df_clean['gender_description']=df_clean.gender.map(d_gender)
df_clean
```

## Feature Engineering

Awesome! Now we want to add a column that is some derivative of an existing column. This is sometimes called **"feature engineering"**. The idea is to create a new variable (aka, column) in our dataset that doesn't currently exist, but is somehow derived from what we already have. 

For instance, what if we want to see which rides were on a weekend? We already have the date of the ride in the `starttime` column, but we want to determine which dates were a Saturday or Sunday. I suspect weekend rides will have different statistical characteristics than weekday rides, so having a column that allows me to easily separate rides into weekend vs weekday rides would be really helpful. 

To do this, we'll use a super helpful pandas method called `.to_datetime()`:

```python
df_clean.starttime = pd.to_datetime(df_clean.starttime)

#now lets check our datatypes to make sure it worked
df_clean.dtypes
```

### Working with Datetime Objects

Perfect. We can now use the `starttime` column as a datetime value instead of a string. This opens up a lot of options - including the ability to figure out if a given date is a weekend or not. 

To start working with datetime objects, let's first import a library called `datetime`. We'll specifically use a method called `.strftime()` to find the day of the week of a given day. We'll pass this method the argument `'%A'`, which tells the datetime library to return the full name of the day, capitalized. Let's try it on the first `starttime` value in our dataset just to see how it works:

```python
import datetime

df_clean.starttime[0].strftime('%A')
```

### Creating a Weekend Indicator Column

The first ride in our dataset (the ride in row 0) was on a Sunday. What I want to do is create a column called `weekend` that has a 1 for rides that were on Saturdays or Sundays, and a 0 for rides on any other day of the week. Let's again use a loop to create our values one row at a time:

```python
weekend=[]

for index, row in df_clean.iterrows():
    #first, let's get the day of the week from the starttime value
    day=row.starttime.strftime('%A')
    #if it is Saturday or Sunday, add a 1 to our weekend list
    if day in ['Saturday', 'Sunday']:
        weekend.append(1)
    #otherwise, add a 0
    else:
        weekend.append(0)
#create a new column with the values from the weekend list
df_clean['weekend']=weekend

#lets see the results!
df_clean.head(10)
```

### Analyzing Weekend vs Weekday Rides

We now have a variable that didn't exist before! We've "engineered a feature"! Now let's do something with it. I wonder if weekend rides are longer on average than weekday rides. Let's find out by grouping on our `weekend` column and finding the mean of our `tripduration` column:

```python
df_clean.groupby(['weekend']).tripduration.mean()
```

## Data Visualization with Seaborn

Turns out, rides on the weekend were 340 seconds longer on average than weekday rides. Now let's try to visualize some of this. We'll use a popular Python plotting library called **Seaborn**. Let's start by first looking at a histogram of our trip durations. We'll use a Seaborn method called `.histplot()`:

```python
import seaborn as sns
sns.histplot(data=df_clean, x='tripduration');
```

### Creating a Density Plot

As suspected, we have a pretty skewed distribution. Let's visualize this as a density plot, which is just a "smoothed" histogram. We do this in Seaborn using a method called `.kdeplot()`, which stands for "kernel density estimation":

```python
sns.kdeplot(data=df_clean, x='tripduration');
```

### Customizing Our Plot

That looks good, but let's play with some of Seaborn's styling features to make this plot our own:

```python
#we'll give our chart a variable name "ax" for "axis". This will allow us to perform further methods on the object after the chart is created
ax = sns.kdeplot(data=df_clean, x='tripduration', fill=True, color='#0072b5ff') #"fill" colors the space between the line and the x-axis. We can set it to whatever color we want, but we'll stick to the default blue

#styling
ax.set(title='CitiBike Trip Durations\nSeptember 2019',
       ylabel='', xlabel='Trip Duration (Seconds)',
      xlim=(0,4000)) #set a title (the \n character creates a line break), clean up our axis labels, and chop x-axis down to start at 0 and stop at 4000
ax.tick_params(left=False, bottom=False) #remove tick labels
med_trip = df_clean.tripduration.median()
ax.axvline(x=med_trip, color='grey', alpha=0.7, linestyle='--') #lets draw a line at the median trip duration

#now lets annotate our plot to call out the median trip duration with an arrow using ax.annotate()
ax.annotate('Median Trip Duration - '+seconds_to_minutes(med_trip), #add what we want 
            xy=(med_trip, 0.00025), xytext=(1000, 0.0005), # xy: The end point of the arrow (arrowhead position); xytext: The start point of the arrow (arrow tail position)
            arrowprops=dict(arrowstyle='->', color='grey', lw=2), #lw: linewidth in pixels
            fontsize=11, color='black')
sns.despine(top=True, right=True) #this line removes the lines from the top and right of the chart area to give it a lighter, more open feel
```

### Creating a Facet Grid

This looks good! Let's try something more complex. Let's create a facet grid using the variables `weekend` and `usertype`. The `usertype` column identifies Citibike riders as:
- **Subscribers**: who pay a yearly membership fee for unlimited rides
- **Customers**: who don't pay the yearly fee, but instead pay a per-minute premium for every ride

I suspect that customers and subscribers will have pretty different median trip durations, and I also suspect that rides on the weekend might be longer than weekday rides. Let's find out!

```python
#instantiate facet gred using the 'weekend' column to represent the rows in the grid and 'usertype' to represent the columns
g = sns.FacetGrid(data=df_clean, col='usertype', row='weekend', aspect=1.5) #use the aspect argument to increase the aspect ratio. Default is 1
#use the .map() method to draw a density plot in each section of the grid
g.map(sns.kdeplot, 'tripduration', fill=True)

#styling
g.set(xlabel='Trip Duration (Seconds)', ylabel='', xlim=(0, 4000)); #set xlimits
g.tick_params(left=False, bottom=False); #remove tick labels

#here's a cool block of code to add a reference line to each plot in the grid.
#first lets find the median trip duration for each group, starting from the top left and going left to right
weekday_subscriber=df_clean.loc[(df_clean.weekend==0)&(df_clean.usertype=='Subscriber')].tripduration.median()
weekday_customer=df_clean.loc[(df_clean.weekend==0)&(df_clean.usertype=='Customer')].tripduration.median()
weekend_subscriber=df_clean.loc[(df_clean.weekend==1)&(df_clean.usertype=='Subscriber')].tripduration.median()
weekend_customer=df_clean.loc[(df_clean.weekend==1)&(df_clean.usertype=='Customer')].tripduration.median()
#now lets create a list with the four values
median_list=[weekday_subscriber, weekday_customer, weekend_subscriber, weekend_customer]
for ax, median in zip(g.axes.flat, median_list):
    ax.axvline(x=median, color='grey', linestyle='--', alpha=0.7)
    #now lets annotate our plot to call out the median trip duration with an arrow using ax.annotate()
    ax.annotate('Median Trip Duration - '+seconds_to_minutes(median), 
                xy=(median, 0.001), xytext=(median+600, 0.0012), 
                arrowprops=dict(arrowstyle='->', color='grey', lw=2), 
                fontsize=11, color='black')
```

## Automating the Entire Workflow

This looks good! To wrap up, let's look at the real benefit of doing data analysis and visualization in Python. We are going to implement several of the things we did in this workshop, but in a **single block of code**. Doing this will allow us to run this entire analysis on a new dataset (for example, a different month of Citibike data) with a single click. Let's get started.

### Our Workflow

We want our script to do the following:
1. Read in the new dataset (Citibike data from November 2019)
2. Remove any outlying values (rides over 2 hours)
3. Find the average and the median trip durations
4. Create a column that maps our gender code (0, 1, and 2) to the gender description
5. Find the top 5 stations that users started rides at during that month and print them, along with how many rides originated there
6. Find the top 5 stations by average trip duration and print the station name and the average trip (in the form minutes:seconds)
7. Create a column to identify weekend rides
8. Create a facetgrid of density plots, faceting on our weekend column and usertype, and export the visual as a .png file
9. Export the cleaned dataset to a .csv file called `citibike_[MMYY]_clean.csv`

We'll also time how long this takes by capturing the time at the beginning and the end and finding the difference. Let's go!

```python
print('importing libraries and setting up functions')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt #we'll use this library to save our visualization as a .png file
import datetime

start_time = datetime.datetime.now() 

def seconds_to_minutes(sec):
    minutes=int(sec/60)
    seconds=round(sec%60)
    if seconds <10:
        seconds='0'+str(seconds)
    return str(minutes)+':'+str(seconds)

print('reading in dataset...')
url='https://raw.githubusercontent.com/nickxscott/INFO658/refs/heads/main/notebooks/data/citibike201911.csv'
df=pd.read_csv(url)
two_hours=2*60*60
print('-----')
print('there are',len(df),'trips in this dataset')
print('-----')
print('cleaning...removing trips longer than two hours..')
df_clean=df.loc[df.tripduration<=two_hours]
removed=len(df)-len(df_clean)
print('nice, we removed ',removed,'rows. Now lets get started.')
print('-----')
avg_trip=df_clean.tripduration.mean()
med_trip=df_clean.tripduration.median()
print('Average trip - ',seconds_to_minutes(avg_trip))
print('Median trip - ',seconds_to_minutes(med_trip))
print('-----')
print('creating gender description column...')
d_gender={0:'unknown',
          1:'male',
          2:'female'}
df_clean.gender=df_clean.gender.map(d_gender)
print('-----')
print('finding top five stations by ride count...')
df_stationCount=df_clean.groupby(['start station name'])['start station name'].count().reset_index(name='count')
df_stationCount=df_stationCount.sort_values(by='count', ascending=False)
for index, row in df_stationCount.head().iterrows():
    print('Station: ', row['start station name'], '; Ride Count: ', row['count'])
print('-----')
print('finding top five stations by average trip duration...')
df_stationDuration=df_clean.groupby(['start station name'])['tripduration'].mean().reset_index(name='average_trip_duration')
df_stationDuration=df_stationDuration.sort_values(by='average_trip_duration', ascending=False)
for index, row in df_stationDuration.head().iterrows():
    print('Station: ', row['start station name'], '; Avg Trip Duration: ', seconds_to_minutes(row['average_trip_duration']))
print('-----')
print('creating weekend identifier column...')
df_clean.starttime = pd.to_datetime(df_clean.starttime)
weekend=[]
for index, row in df_clean.iterrows():
    day=row.starttime.strftime('%A')
    if day in ['Saturday', 'Sunday']:
        weekend.append(1)
    else:
        weekend.append(0)
df_clean['weekend']=weekend
print('-----')
print('creating facetgrid...')
g = sns.FacetGrid(data=df_clean, col='usertype', row='weekend', aspect=1.5)
g.map(sns.kdeplot, 'tripduration', fill=True)
g.set(xlabel='Trip Duration (Seconds)', ylabel='', xlim=(0, 4000));
g.tick_params(left=False, bottom=False); 
weekday_subscriber=df_clean.loc[(df_clean.weekend==0)&(df_clean.usertype=='Subscriber')].tripduration.median()
weekday_customer=df_clean.loc[(df_clean.weekend==0)&(df_clean.usertype=='Customer')].tripduration.median()
weekend_subscriber=df_clean.loc[(df_clean.weekend==1)&(df_clean.usertype=='Subscriber')].tripduration.median()
weekend_customer=df_clean.loc[(df_clean.weekend==1)&(df_clean.usertype=='Customer')].tripduration.median()
median_list=[weekday_subscriber, weekday_customer, weekend_subscriber, weekend_customer]
for ax, median in zip(g.axes.flat, median_list):
    ax.axvline(x=median, color='grey', linestyle='--', alpha=0.7)
    ax.annotate('Median Trip Duration - '+seconds_to_minutes(median), 
                xy=(median, 0.001), xytext=(median+600, 0.0012), 
                arrowprops=dict(arrowstyle='->', color='grey', lw=2), 
                fontsize=11, color='black')
#get month and year of dataset to create a unique filename for our plot
print('-----')
print('saving visual as .png file...')
month=str(df_clean.starttime[0].month)
year=str(df_clean.starttime[0].year)
plt.savefig('citbike_'+year+month+'.png', bbox_inches='tight')
print('-----')
print('saving clean dataset as .csv file')
df_clean.to_csv('citibike_'+year+month+'_clean.csv', index=False)
print('-----')
end_time = datetime.datetime.now() 
time_difference = end_time - start_time
print(f'All done! That took: {time_difference} seconds')
```
